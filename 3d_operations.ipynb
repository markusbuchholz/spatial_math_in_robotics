```python
#rotate about Z = 30, after around Y=20 and X=10
#rotate by 30° about the Z-axis (yaw), then 20° about the Y-axis (pitch) and then 10° about the X-axis – this is the ZYX roll-pitch yaw convention
SE3.RPY([10, 20, 30], unit='deg')
```

```python
Ha = SE3(0,0,0)
Hb = SE3(2.8,0.5, 0)
Hc = SE3.Tx(1.0) * SE3.Ty(1.0) * SE3.Tz(2) * SE3.Rz(-45, 'deg')* SE3.Ry(45, 'deg')* SE3.Rx(45, 'deg')
Hd = SE3.Tx(2) * SE3.Ty(3) * SE3.Tz(3) * SE3.Rz(60, 'deg')* SE3.Ry(60, 'deg')* SE3.Rx(-60, 'deg')
He = SE3.Tx(1) * SE3.Ty(2) * SE3.Tz(-2.0) * SE3.Rz(30, 'deg')* SE3.Ry(30, 'deg')* SE3.Rx(30, 'deg')

###

Hde = Hd*He
Hde
```

```python
Ha.plot(frame='A', color='black', dims=[0,4])
Hb.plot(frame='B', color='blue', dims=[0,4])
Hc.plot(frame='C', color='green', dims=[0,4])
Hd.plot(frame='D', color='navy', dims=[0,4])
Hde.plot(frame='E', color='red', dims=[0,4])
```

```python
Hce = (Ha* Hb * Hb * Hc).inv() * Ha*Hd * Hd*He;
Hce
```
